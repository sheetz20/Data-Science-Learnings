```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris.columns
```

```python
iris['species'].value_counts()
```

```python
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.show()
```

```python
sns.set_style('whitegrid')
g = sns.FacetGrid(iris,hue='species',size=4)
g.map(plt.scatter,'sepal_length','sepal_width')
g.add_legend()
plt.show()
```

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris,hue='species',size=3)
plt.show()
```

```python
#New Inputs to distinguish into different Species
petal_length = float(input('enter new petal length '))
petal_width = float(input('enter new pental width '))
if petal_width < 1 and petal_length < 2:
    print('setosa')
elif 1<petal_width<2 and 3<petal_length<5:
    print('Versicolor')
else:
    print('Virginica')
```
